# 5. 데이터셋 이상치 제거 및 스케일링

## 5-1 IQR(Inter Quantile Range) 활용한 이상치 제거

```python
def remove_outlier(dataframe, remove_col):
    dff = dataframe
    for k in remove_col:
        level_1q = dff[k].quantile(0.25)
        level_3q = dff[k].quantile(0.75)
        IQR = level_3q - level_1q
        rev_range = 1.5  # 제거 범위 조절 변수
        dff = dff[(dff[k] <= level_3q + (rev_range * IQR)) & (dff[k] >= level_1q - (rev_range * IQR))]
#         dff = dff.reset_index(drop=True)
    return dff
```

```python
remove_col = [ 'slag', 'water', 'superplastic', 'fineagg', 'age', 'cement/water(%)', 'coarseagg/fineagg', 'water/age']

concrete_df_reout = remove_outlier(concrete_df, remove_col)
concrete_df_reout
```

# 6. 1차 알고리즘 적용 (하이퍼 파라미터 X)

## 6-1 다중 선형 회귀, Ridge, Lasso, ElasticNet 적용

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MaxAbsScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import RobustScaler
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.linear_model import ElasticNet
from sklearn.metrics import mean_squared_error

# 데이터셋 불러오기
X = concrete_df.iloc[:, [0,1,2,3,4,5,6,7,9]]
y = concrete_df.iloc[:, [8]]

RS_ARR = [7,77,777]
RMSE_SUM = 0
df_data = []

models = [LinearRegression(), Ridge(), Lasso(), ElasticNet()]
model_name = ['LinearRegression', 'Ridge', 'Lasso', 'ElasticNet']

for md in models:
    data = []
    
    for RS in RS_ARR:
        # train, test 데이터 구분
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=RS)

        x_train_poly = PolynomialFeatures().fit_transform(X_train)
        x_test_poly = PolynomialFeatures().fit_transform(X_test)
        
        scaler = RobustScaler()
        scaler.fit(x_train_poly)
        X_train_scaled = scaler.transform(x_train_poly)
        X_test_scaled = scaler.transform(x_test_poly)
        
        model = md.fit(X_train_scaled, y_train)
        y_pred = model.predict(X_test_scaled)

        train_score = model.score(X_train_scaled, y_train)
        test_score =model.score(X_test_scaled, y_test)

        model_RMSE = (np.sqrt(mean_squared_error(y_test, y_pred)))

        RMSE_SUM += model_RMSE

        data.append(train_score)

        data.append(test_score)

        data.append(model_RMSE)
    
    data.append(RMSE_SUM / 3)
    df_data.append(data)
    RMSE_SUM = 0
    
results_df = pd.DataFrame(df_data, columns = ['7_Train_Score', '7_Test_Score', '7_RMSE',
                                                    '77_Train_Score', '77_Test_Score', '77_RMSE',
                                                    '777_Train_Score', '777_Test_Score', '777_RMSE', 'Average_RMSE'])
results_df.insert(0, 'Model_Name', model_name)
results_df
```

## 7-1-1 Random_state = 7, Hyper-parameter tuning

```python
# Optuna Libraries
import optuna
from optuna import Trial
from optuna.samplers import TPESampler

X = concrete_df.iloc[:, [0,1,2,3,4,5,6,7,9]]
y = concrete_df.iloc[:, [8]]
y = y.values.ravel()

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=7)

# random sampler
sampler = TPESampler(seed=10)

# define function
def objective(trial):
    cbrm_param = {
        'iterations':trial.suggest_int("iterations", 100, 25000),
        'od_wait':trial.suggest_int('od_wait', 500, 2300),
        'learning_rate' : trial.suggest_float('learning_rate',0.01, 1),
        'reg_lambda': trial.suggest_float('reg_lambda',1e-5,100),
        'subsample': trial.suggest_float('subsample',0,1),
        'random_strength': trial.suggest_float('random_strength',1,50),
        'depth': trial.suggest_int('depth',1, 15),
        'min_data_in_leaf': trial.suggest_int('min_data_in_leaf',1,30),
        'leaf_estimation_iterations': trial.suggest_int('leaf_estimation_iterations',1,15),
        'bagging_temperature' :trial.suggest_float('bagging_temperature', 0.01, 100.00),
        'colsample_bylevel':trial.suggest_float('colsample_bylevel', 0.1, 1.0),
    }

    # Generate model
    model_cbrm = CatBoostRegressor(**cbrm_param)
    model_cbrm = model_cbrm.fit(X_train, y_train, eval_set=[(X_test, y_test)], 
                           verbose=0, early_stopping_rounds=25)
                           
    # 평가지표 원하는 평가 지표가 있을 시 바꾸어 준다.
    MSE = mean_squared_error(y_test, model_cbrm.predict(X_test))
    return MSE

optuna_cbrm = optuna.create_study(direction='minimize', sampler=sampler)
optuna_cbrm.optimize(objective, n_trials=50)
```

## 7-1-2 가장 높은 튜닝 점수와 파라미터 값 추출

```python
cbrm_trial = optuna_cbrm.best_trial
cbrm_trial_params = cbrm_trial.params
print('Best Trial: score {},\nparams {}'.format(cbrm_trial.value, cbrm_trial_params))
```
